````python
# batch generator
def get_batches(dataset, batch_size):
    X, Y = dataset
    n_samples = X.shape[0]
        
    # Shuffle at the start of epoch
    indices = np.arange(n_samples)
    np.random.shuffle(indices)
    
    for start in range(0, n_samples, batch_size):
        end = min(start + batch_size, n_samples)
        
        batch_idx = indices[start:end]
    
        yield X[batch_idx], Y[batch_idx]
````

````python
from typing import Tuple, List
from tqdm.notebook import tqdm
import seaborn as sns

%config InlineBackend.figure_format="retina"
sns.set(style='darkgrid', font_scale=1.4,palette='Set2')
np.random.seed(42)
````

````python
N_EPOCH = 30
BATCH_SIZE = 128
````

````python
def build_default_image_classifier_nn(custom_activation, with_batchnorm: bool, 
                                      with_dropout: bool = False, pipeline = None):
    if pipeline is None:
        pipeline = [28 * 28, 64, 10]
    net = Sequential()
    net.add(Flatten())
    for i in range(0, len(pipeline) - 1):
        net.add(Linear(pipeline[i], pipeline[i + 1]))
        
        if with_batchnorm:
            net.add(BatchNormalization(0.3))
            net.add(ChannelwiseScaling(pipeline[i + 1]))
        if with_dropout:
            net.add(Dropout(0.2))
            
        if i != len(pipeline) - 2:
            net.add(custom_activation())

    net.add(LogSoftMax())
    return net
````

````python
class ImageClassifierNet:
    all_activations = {
        'ReLU': ReLU,
        'ELU': ELU,
        'LeakyReLU': LeakyReLU,
        'SoftPlus': SoftPlus
    }
    all_optimizers = {
        'adam': adam_optimizer,
        'sgd_momentum': sgd_momentum,
    }
    
    def __init__(self, activation: str, optimizer: str, optimizer_config: dict,n_epoch: int = N_EPOCH, 
                 batch_size: int = BATCH_SIZE, with_batchnorm = False, with_dropout = False, 
                 pipeline: List[int] = None):
        assert activation in self.all_activations.keys(), 'Not specified activation function'
        assert optimizer in self.all_optimizers.keys(), 'Not specified optimizer'
        
        self.activation_name = activation
        self.with_batchnorm = with_batchnorm

        self.optimizer = self.all_optimizers[optimizer]
        self.optimizer_config = optimizer_config
        
        self.net = build_default_image_classifier_nn(
            self.all_activations[activation], 
            with_batchnorm,
            with_dropout,
            pipeline
        )
        self.criterion = ClassNLLCriterion()

        self.n_epoch = n_epoch
        self.batch_size = batch_size

        self.train_loss = None
        self.val_loss = None
        

    def fit(self, dataset, verbose: bool = True) -> Tuple[List[float], List[float]]:
        X_train, y_train, X_val, y_val = dataset
        optimizer_state = {}
        train_loss, val_loss = [], []
        
        for epoch in range(self.n_epoch):
            self.net.train()
            cur_train_loss = []
            for x_batch, y_batch in tqdm(get_batches((X_train, y_train), self.batch_size)):
                self.net.zeroGradParameters()
                predictions = self.net.forward(x_batch)
                loss = self.criterion.forward(predictions, y_batch)
                
                dp = self.criterion.backward(predictions, y_batch)
                self.net.backward(x_batch, dp)
                self.optimizer(self.net.getParameters(), self.net.getGradParameters(), 
                               self.optimizer_config,optimizer_state)      
                cur_train_loss.append(loss)
            train_loss.append(np.mean(cur_train_loss))
                
            
            self.net.evaluate()
            cur_val_loss = []
            for x_batch_val, y_batch_val in get_batches((X_val, y_val), self.batch_size):
                y_pred = self.net.forward(x_batch_val)
                loss = self.criterion.forward(y_pred, y_batch_val)
                cur_val_loss.append(loss)

            val_loss.append(np.mean(cur_val_loss))
            
            if verbose:
                display.clear_output(wait=True)
                print(f'Epoch: {epoch + 1}, train loss: {train_loss[-1]:.3f},' + \
                      f'validation loss {val_loss[-1]:.3f}')
        self.train_loss = train_loss
        self.val_loss = val_loss
        return self
    
    def predict(self, X) -> np.array:
        self.net.evaluate()
        pred = self.net.forward(X)
        return np.argmax(pred, axis=-1)
    
    def accuracy_score(self, X, y) -> float:
        pred = self.predict(X)
        return np.mean(pred == np.argmax(y, axis=-1))
    

    def __repr__(self):
        return self.net.__repr__()
    
    def get_title(self) -> str:
        batchnorm =  ' with' if self.with_batchnorm else ' without'
        return self.activation_name + ' activation' + batchnorm + ' batchnorm'
    
    def plot_train_curve(self):
        assert self.train_loss is not None, 'Net is not trained yet'
        
        fig, axis = plt.subplots(figsize=(15, 8))
        plt.title(self.get_title(), fontsize=18)
        
        grid_epoch = np.arange(self.n_epoch, dtype=np.int32) + 1
        plt.plot(grid_epoch, self.train_loss, label='Train loss')
        plt.plot(grid_epoch, self.val_loss, label='Validation loss')

        axis.set_xlabel('Epoch', fontsize=16)
        axis.set_ylabel('Loss value', fontsize=16)
        axis.legend()
        plt.show()
        return
    
    def compare_with_batchnorm(self, other):
        assert not self.with_batchnorm, 'Calling function on batchnorm net'
        assert other.with_batchnorm, 'Other net doesn\'t contain a batchnorm layer'
        assert other.n_epoch == self.n_epoch
                
        fig, axis = plt.subplots(figsize=(15, 8))
        plt.title(self.activation_name + ' batchnorm comparison', fontsize=18)
        
        grid_epoch = np.arange(self.n_epoch, dtype=np.int32) + 1
        
        plt.plot(grid_epoch, self.train_loss, label='Train loss without batchnorm')
        plt.plot(grid_epoch, self.val_loss, label='Validation loss without batchnorm')
        
        plt.plot(grid_epoch, other.train_loss, label='Train loss with batchnorm', linestyle='dashdot')
        plt.plot(grid_epoch, other.val_loss, label='Validation loss with batchnorm', linestyle='dashdot')

        axis.set_xlabel('Epoch', fontsize=16)
        axis.set_ylabel('Loss value', fontsize=16)
        axis.legend()
        plt.show()
        return
````

````python
import torch
import torch.nn.functional as F
from torch import nn
````

````python
import mnist
X_train, y_train, X_val, y_val, X_test, y_test = mnist.load_dataset()
````

````python
class MnistDataset(torch.utils.data.Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y
        
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        return {"sample": torch.tensor(self.X[idx, :], dtype=torch.float), 
                "target": torch.tensor(self.y[idx], dtype=torch.long)}
````

````python
train_ds = MnistDataset(X_train, y_train)
val_ds = MnistDataset(X_val, y_val)
test_ds = MnistDataset(X_test, y_test)
````

````python
train_dataloader = torch.utils.data.DataLoader(
    train_ds, 
    batch_size=BATCH_SIZE,
    shuffle=True, 
    num_workers=1
)
val_dataloader = torch.utils.data.DataLoader(
    val_ds, 
    batch_size=BATCH_SIZE,
    shuffle=True, 
    num_workers=1
)
test_dataloader = torch.utils.data.DataLoader(
    test_ds, 
    batch_size=BATCH_SIZE,
    shuffle=False, 
    num_workers=1
)
````

````python
model = nn.Sequential( 
    nn.Flatten(),
    nn.Linear(784, 256),
    nn.BatchNorm1d(256),
    nn.Dropout(0.2),
    nn.ReLU(),
    nn.Linear(256, 128),
    nn.BatchNorm1d(128),
    nn.Dropout(0.2),
    nn.ReLU(),
    nn.Linear(128, 64),
    nn.BatchNorm1d(64),
    nn.Dropout(0.2),
    nn.ReLU(),
    nn.Linear(64, 10),
    nn.BatchNorm1d(10),
    nn.Dropout(0.2),
    nn.LogSoftmax()
)
````

````python
nll_loss = nn.NLLLoss()
optimizer = torch.optim.Adam(model.parameters(), betas=(0.3, 0.4), lr=3*1e-3, eps=1e-5)
````

````python
for epoch in range(15):
    model.train()
    train_loss = []
    for batch in tqdm(train_dataloader):
        x_batch, y_batch = batch['sample'], batch['target']
        y_pred = model(x_batch)  
        loss = nll_loss(y_pred, y_batch)
        loss.backward()   
        optimizer.step() 
        optimizer.zero_grad()
        train_loss.append(loss)
    
    model.eval()
    val_loss = []
    with torch.no_grad():                          
        for batch in tqdm(val_dataloader):
            x_batch, y_batch = batch['sample'], batch['target']
            y_pred = model(x_batch) 
            loss = nll_loss(y_pred, y_batch)
            val_loss.append(loss)
        display.clear_output(wait=True)
        print(f"Epoch: {epoch + 1}, train loss: {np.mean(train_loss)}, validation loss: {np.mean(val_loss)}")
````

````python
predicts = []
model.eval()
with torch.no_grad():
        for batch in test_dataloader:
            sample, target = batch['sample'], batch['target']
            y_pred = model(sample)
            predicts.extend((torch.argmax(y_pred, dim=-1) == target).numpy().tolist())

print('Accuracy:', np.mean(predicts))
````
